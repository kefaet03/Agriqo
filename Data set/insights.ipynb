```python
import os
import random
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import warnings
warnings.filterwarnings("ignore")

data_dir = data_dir = "E:\\He_is_enough03 X UniqoXTech X Dreams\\Click_here\\Artificial Intelligence\\ML - AGRIQO\\Data set\\dataset01.csv"

data = pd.read_csv(data_dir)
```

```python
import seaborn as sns

plt.figure(figsize=(15, 6))
sns.countplot(data['label'], order=data['label'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Label Distribution Before Augmentation")
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))


sns.boxplot(data=data, x='label', y='Mean Temp. (*C)', palette="pastel")
plt.title("Temperature-Label")
plt.xlabel("Label")
plt.ylabel("temp")

plt.xticks(rotation=90, ha='right')

plt.tight_layout()
plt.show()
```

```python
cols = ['Rainfall (mm)','Mean Temp. (*C)', 'RHmean (%)', 'SShr (hrs)', 'WD (deg)', 'WS (Km/hr)']

for col in cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))


sns.boxplot(data=data, x='label', y='Mean Temp. (*C)', palette="pastel")
plt.title("Temperature-Label")
plt.xlabel("Label")
plt.ylabel("temp")

plt.xticks(rotation=90, ha='right')

plt.tight_layout()
plt.show()
```
