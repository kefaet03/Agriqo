```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import warnings
warnings.filterwarnings("ignore")

data_dir = data_dir = "E:\\He_is_enough03 X UniqoXTech X Dreams\\Click_here\\Artificial Intelligence\\ML - AGRIQO\\Data set\\Agriqo(slider1) - Sheet1.csv"

data = pd.read_csv(data_dir)

data.head()
```

```python
data['Month']=data['month(chara)'].astype(int)
data['Week'] = ((data['month(chara)']-data['Month'])*4+1).astype(int)
import calendar
data['Month'] = data['Month'].apply(lambda x: calendar.month_name[x])

data.head()
```

```python
data['Agricultural Zone']=data['Agricultural zone']
data['Rainfall (mm)'] = data['rainfall (mm)']
data['Mean Temp. (*C)'] = data['temperature(avg)']
data['RHmean (%)'] = data['humidity(avg)']
data['SShr (hrs)'] = data['sunlight(hour)']
data['WD (deg)'] = data['direction of wind(deg)']
data['WS (Km/hr)'] = data['velocity of wind(km/h)']

data.head()
```

```python
data.columns
```

```python
col = ['Agricultural zone', 'rainfall (mm)','temperature(avg)', 'humidity(avg)', 'sunlight(hour)',
       'direction of wind(deg)', 'velocity of wind(km/h)','month(chara)','Unnamed: 9','Unnamed: 10',
       'Unnamed: 11','Unnamed: 12']
data.drop(columns=col, axis=1, inplace=True)

data.head()
```

```python
data = data[['Agricultural Zone','Month','Week', 'Rainfall (mm)', 'Mean Temp. (*C)','RHmean (%)', 'SShr (hrs)', 'WD (deg)', 'WS (Km/hr)','label']]
data.head()
```

```python
data.to_csv('dataset01.csv',index=False)
```
